```python
X_cat = X.select_dtypes(["O","int"])
X_cat.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
print(X.shape, X_train.shape, X_test.shape, y_test.shape)
```

Scale the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import pickle
filename = 'stdscaler.sav'
pickle.dump(scaler,open(filename,'wb'))
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
search_results = {}
```

```python
tree_model=DecisionTreeClassifier()

parameters = {
    'criterion': ('gini', 'entropy'),
    'max_depth': (3, 40),
    'min_samples_split': (3, 40),
    'min_samples_leaf': (3, 40)
}

search = RandomizedSearchCV(
    tree_model,
    param_distributions=parameters,
    scoring="roc_auc",            # Metric to optimize
    cv=5,                         # cross-validation
    refit=True,
    n_jobs=-1,                    # Use all th cpu capacity
    return_train_score=True,
    n_iter=6                  # Number of iterations
)
```

```python
search.fit(X_train, y_train)
```

```python
search_results[tree_model] = search.best_params_
# Entrena un modelo final con los mejores hiperparÃ¡metros
best_tree_model = DecisionTreeClassifier(**search_results[tree_model])
best_tree_model.fit(X_train, y_train)
```

```python
X_test_prediction = search.predict(X_test)
```

```python
print(classification_report(y_test,X_test_prediction))
```

```python
print(confusion_matrix(y_test,X_test_prediction))
```

```python
from sklearn.metrics import accuracy_score
#training_data_accuracy = accuracy_score(X_train_prediction, y_train) #tengo que predecir priero en train??
#training_data_accuracy
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
test_data_accuracy
```

```python
import pickle
filename = 'tree_trained_model.sav'
pickle.dump(best_tree_model,open(filename,'wb'))
```

Random Forest Classifier

```python
model = RandomForestClassifier(n_jobs=-1)

parameters = {
    'n_estimators': (10, 100),         # NÃºmero de Ã¡rboles en el bosque
    'max_depth': (10, 100),            # Profundidad mÃ¡xima de los Ã¡rboles
    'min_samples_split': (2, 20),      # NÃºmero mÃ­nimo de muestras requeridas para dividir un nodo interno
    'min_samples_leaf': (1, 10)        # NÃºmero mÃ­nimo de muestras requeridas en un nodo hoja
}

search = RandomizedSearchCV(
    model,
    param_distributions=parameters,
    scoring="roc_auc",            # Metric to optimize
    cv=5,                         # cross-validation
    refit=True,
    n_jobs=-1,                    # Use all th cpu capacity
    return_train_score=True,
    n_iter=6                  # Number of iterations
)
```

```python
search.fit(X_train, y_train)
```

```python
search_results[model] = search.best_params_
# Entrena un modelo final con los mejores hiperparÃ¡metros
best_rfo_model = RandomForestClassifier(**search_results[model])
best_rfo_model.fit(X_train, y_train)
```

```python
X_test_prediction = search.predict(X_test)
```

```python
print(classification_report(y_test,X_test_prediction))
```

```python
print(confusion_matrix(y_test,X_test_prediction))
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
test_data_accuracy
```

```python
import pickle
filename = 'rfo_trained_model.sav'
pickle.dump(best_rfo_model,open(filename,'wb'))
```
